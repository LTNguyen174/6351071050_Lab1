Câu 1:

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'tb_tracking.xlsx'
df = pd.read_excel(file_path, header=None)

df.columns = ['id', 'device_id', 'co_level', 'humidity', 'temperature', 'time']

# Chuyển đổi cột 'time' sang dạng datetime để xử lý thời gian
df['time'] = pd.to_datetime(df['time'])
```

câu 2:

```python
df = df.dropna(subset=['time']) # Loại bỏ dòng lỗi thời gian
df_filtered = df[df['time'] >= '2018-01-03'].copy()
```

câu 3:

```python
df_filtered['minutes'] = df_filtered['time'].dt.hour * 60 + df_filtered['time'].dt.minute

# Xem trước dữ liệu
print("Dữ liệu sau khi xử lý:")
print("Lý Thanh Nguyên - 6351071050\n")
print(df_filtered.head())
```

Câu 4:

```python
plt.figure(figsize=(18, 6))
print("Lý Thanh Nguyên - 6351071050\n")

# Đồ thị 1: Nồng độ CO
plt.subplot(1, 3, 1)
plt.scatter(df_filtered['minutes'], df_filtered['co_level'], alpha=0.5, c='blue')
plt.title('Nồng độ CO theo thời gian (phút)')
plt.xlabel('Số phút trong ngày')
plt.ylabel('Nồng độ CO')

# Đồ thị 2: Độ ẩm
plt.subplot(1, 3, 2)
plt.scatter(df_filtered['minutes'], df_filtered['humidity'], alpha=0.5, c='green')
plt.title('Độ ẩm theo thời gian (phút)')
plt.xlabel('Số phút trong ngày')
plt.ylabel('Độ ẩm (%)')

# Đồ thị 3: Nhiệt độ
plt.subplot(1, 3, 3)
plt.scatter(df_filtered['minutes'], df_filtered['temperature'], alpha=0.5, c='red')
plt.title('Nhiệt độ theo thời gian (phút)')
plt.xlabel('Số phút trong ngày')
plt.ylabel('Nhiệt độ')

plt.tight_layout()
plt.show()
```
